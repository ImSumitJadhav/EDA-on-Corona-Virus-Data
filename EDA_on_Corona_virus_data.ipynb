#<font color='red'> **Deal with NaN values "-"**

```python
#Fill NaN values with
df1=df1.fillna("-")
df2=df2.fillna("-")
df3=df3.fillna("-")
```

#<font color='red'> **Visualization**

#<font color='red'> **1)which country has most Number of confirmed cases,death cases and recover cases**

```python
df1.head()
```

```python
confirmed_on_day_one=pd.DataFrame(df1.groupby("Country/Region")["Confirmed"].sum()).reset_index() #<<<<<<<<<<<<<<
confirmed_on_day_one=confirmed_on_day_one.nlargest(10,"Confirmed")
confirmed_on_day_one
```

```python
#plot graph
plt.bar(confirmed_on_day_one["Country/Region"],confirmed_on_day_one["Confirmed"])
plt.xlabel("Country/Region")
plt.xticks(rotation ='vertical')
plt.ylabel("Confirmed cases in millions")
```

```python
death_on_day_one=pd.DataFrame(df1.groupby("Country/Region")["Deaths"].sum()).reset_index()
death_on_day_one=death_on_day_one.nlargest(10,"Deaths")
death_on_day_one
```

```python
#plot graph
plt.bar(death_on_day_one["Country/Region"],death_on_day_one["Deaths"],color="r")
plt.xlabel("Country/Region")
plt.xticks(rotation ='vertical')
plt.ylabel("Deaths cases in millions")
```

```python
recover_on_day_one=pd.DataFrame(df1.groupby("Country/Region")["Recovered"].sum()).reset_index()
recover_on_day_one=recover_on_day_one.nlargest(10,"Recovered")
recover_on_day_one
```

```python
#plot graph
plt.bar(recover_on_day_one["Country/Region"],recover_on_day_one["Recovered"],color="c")
plt.xlabel("Country/Region")
plt.xticks(rotation ='vertical')
plt.ylabel("Recovered cases in millions")
```

#<font color='red'> **2)At which time period (Dates) this virus as its peak**

```python
l=["Confirmed","Deaths","Recovered"]      #list of key column names
```

```python
#plot graph by key column names
for i in l:
  fig,axes=plt.subplots(1,figsize=(12,8))
  a=pd.DataFrame(df1.groupby("ObservationDate")[i].sum()).reset_index()
  a=a.nlargest(15,i)
  plt.bar(a["ObservationDate"],a[i],color="c")
  plt.xticks(rotation ='vertical')
  plt.xlabel("ObservationDate")
  plt.ylabel(i+" in millions")
```

#<font color='red'> **3)Information about India and Corona virus**

```python
#dataframe of india
india_info=df1[df1["Country/Region"]=="India"]
india_info.head(10)
```

```python
#store total values of confirmed,death and recover
ind=[]
for n in l:
  ind_c=india_info[n].sum()
  ind.append(ind_c)

ind  
```

```python
#ratio of "Confirmed","Deaths","Recovered"
lab=["Confirmed","Deaths","Recovered"]
plt.figure(figsize=(17,10))
plt.pie(ind,labels=lab,colors=["r","b","orange"],shadow=True)
```

```python
#state wise coparision
ind_st=pd.DataFrame(india_info.groupby("Province/State")["Confirmed"].mean()).reset_index()[1:]
ind_st=ind_st.nlargest(10,"Confirmed")
ind_st
```

```python
#plot graph to show which state suffered most
plt.bar(ind_st["Province/State"],ind_st["Confirmed"])
plt.xticks(rotation ='vertical')
plt.xlabel("Province or State in India")
plt.ylabel("Confirmed cases in millions")
```

#<font color='red'> **4)At the end of date 29th May 2021 which country suffered most**

```python
df2.head()
```

```python
tol_cases=pd.DataFrame(df2.groupby("Country/Region")["5/29/21"].sum()).reset_index()
top_ten=tol_cases.nlargest(10,"5/29/21")     #top ten
last_ten=tol_cases.nsmallest(10,"5/29/21")   #last ten
```

```python
top_ten
```

```python
#Number of confirmed cases in millions vs Country/Region
sns.barplot(top_ten["Country/Region"],top_ten["5/29/21"])
plt.xticks(rotation='vertical')
plt.ylabel("Number of confirmed cases in millions")
```

```python
last_ten
```

```python
#Number of confirmed cases vs Country/Region
sns.barplot(last_ten["Country/Region"],last_ten["5/29/21"])
plt.xticks(rotation='vertical')
plt.ylabel("Number of confirmed cases")
```
